# Euler's Equations
Euler's equations are given by:

$$
\frac{\partial U}{\partial t}+\frac{\partial F(U)}{\partial x}=0 \quad \text { with } \quad U=\left[\begin{array}{c}
\rho \\
\rho u \\
 E
\end{array}\right] \quad \text { and } \quad F(U)=\left[\begin{array}{c}
\rho u \\
\rho u^2+p \\
(E+p) u
\end{array}\right].
$$

Our closure will be a caloric equation of state given by,

$$
p=(\gamma-1) \rho\left[E-\frac{1}{2} u^2\right] .
$$


## Jacobian

The Jacobian is,

$$
A(U)=\left[\begin{array}{ccc}
0 & 1 & 0 \\
-\frac{1}{2}(\gamma-3)\left(\frac{u_2}{u_1}\right)^2 & (3-\gamma)\left(\frac{u_2}{u_1}\right) & \gamma-1 \\
-\frac{\gamma u_2 u_3}{u_1^2}+(\gamma-1)\left(\frac{u_2}{u_1}\right)^3 &\frac{\gamma u_3}{u_1}-\frac{3}{2}(\gamma-1)\left(\frac{u_2}{u_1}\right)^2 & \gamma\left(\frac{u_2}{u_1}\right)
\end{array}\right] ,
$$
with eigen values,
$$
\lambda_1=u-a, \lambda_2=u, \lambda_3=u+a
$$ 
and the corresponding right eigenvectors,
$$
\begin{aligned}
&K^{(1)}=\left[\begin{array}{c}
1 \\
u-a \\
H-u a
\end{array}\right], K^{(2)}=\left[\begin{array}{c}
1 \\
u \\
\frac{1}{2} u^2
\end{array}\right], K^{(3)}=\left[\begin{array}{c}
1 \\
u+a \\
H+u a
\end{array}\right]
\end{aligned}
$$
where $a$ is the sound speed defined by 
$$
a=\left.\sqrt{\partial_\rho p}\right|_s
$$
at fixed entropy $s$.

$\lambda_2$ is associated with a linear wave since,
$$
\nabla_U \lambda_2 \cdot K^{(2)}=[-u / \rho, 1 / \rho, 0] \cdot\left[\begin{array}{c}
1 \\
u \\
\frac{1}{2} u^2
\end{array}\right]=0
$$
while the others are "genuinely non-linear".


## Generalized Riemann Invariants
[see this stackexchange  post](https://math.stackexchange.com/questions/4621142/anyone-able-to-derive-the-following-riemann-invariant-relations)

Consider the the variables, $Q$, conserved along integral curves of $K$. 
Then for,
$$
\mathrm{d}_\xi U = K,
$$
and
$$
\mathrm{d}_\xi Q(U) = \nabla Q(U)\cdot K = 0
$$

we can use the [Lagrange Charpitt](https://en.wikipedia.org/wiki/Method_of_characteristics) equations for our system to determine $Q$,
$$
\mathrm{d}\xi=\frac{\mathrm{d} u_1}{k_1}=\frac{\mathrm{d}u_2}{k_2}=\frac{\mathrm{d} k_3}{k_3}=\cdots
$$

The parameter $\xi$ changes *across* characteristic curves since,

$$
\begin{aligned}
    \partial_t U +A(u)\partial_x U
        &=0\\
    \implies (\partial_t\xi +A(u)\partial_x \xi)\partial_\xi U
        &=0\\
    \implies  (\partial_t\xi +A(u)\partial_x \xi) K
        &=0
\end{aligned}
$$


## Wave Structure

An example wave pattern is shown below with the corresponding characteristic directions associated with each eigen value



The rarefaction wave satisfy
$$
\lambda_1\left(\mathbf{U}_{\mathrm{L}}\right) \leq \lambda_1\left(\mathbf{U}_{* \mathrm{~L}}\right)
$$

indicating that the left characteristics move slower than the right.

The shock wave satisfy,
$$
\lambda_3\left(\mathbf{U}_{* \mathrm{R}}\right)>S_3>\lambda_3\left(\mathbf{U}_{\mathrm{R}}\right)
$$
indicating that the left characteristics run into the right. $S_3$ is the speed of the 3-shock.
Finally, consistency of the linear eigen value requires

$$
\lambda_2\left(\mathbf{U}_{* \mathrm{~L}}\right)=\lambda_2\left(\mathbf{U}_{* \mathrm{R}}\right)=S_2,
$$
where $S_2$ i the speed of the contact wave.

#### Contact Discontinuity

Consider the the variables, $Q$, conserved along integral curves of $K$. 
Then for,
$$
\mathrm{d}_\xi U = K.
$$
and
$$
\mathrm{d}_\xi Q(U) = \nabla Q(U)\cdot K = 0
$$
We can use the [Lagrange Charpitt](https://en.wikipedia.org/wiki/Method_of_characteristics) equations for our system to determine $Q$,
$$
\mathrm{d}\xi=\frac{\mathrm{d} \rho}{1}=\frac{\mathrm{d}(\rho u)}{u}=\frac{\mathrm{d} E}{\frac{1}{2} u^2}
$$
which imply 
$$
\mathrm{d}_\xi p=0\rightarrow p=\text { constant }
$$
and
$$
\mathrm{d}_\xi u=0\rightarrow u=\text { constant }
$$

#### Rarefaction Waves
The Lagrange-Charpit equations for $\lambda_{1/3}$ are
$$
\frac{\mathrm{d}\rho}{1} = \frac{\mathrm{d}u}{\mp a/\rho} = \frac{\mathrm{d}s}{0}
$$

The Generalised Riemann Invariants accross rarefactions waves for $\lambda_1$ and $\lambda_3$ are
$$
\lambda_1=u-a\rightarrow\begin{cases}
I_L(u,a)=u+\frac{2a}{\gamma-1}
    & = \text{constant}\\
    &s = \text{constant}
\end{cases}
$$
$$
\lambda_3=u_a\rightarrow\begin{cases}
I_R(u,a)=u-\frac{2a}{\gamma-1}
    & = \text{constant}\\
    &s = \text{constant}
\end{cases}
$$

#### Shock Waves

We will use the Rankine-Hugonoit Conditions, which state
$$
\Delta F = S_i\Delta U
$$
where $S_i$ is the wave solution speed associated with $\lambda_i$, and
$$
\Delta \mathbf{U} \equiv \mathbf{U}_{\mathrm{R}}-\mathbf{U}_{*}, \Delta \mathbf{F} \equiv \mathbf{F}_{\mathrm{R}}-\mathbf{F}_{*}, \mathbf{F}_{*}=\mathbf{F}\left(\mathbf{U}_{*}\right), \mathbf{F}_{\mathrm{R}}=\mathbf{F}\left(\mathbf{U}_{\mathrm{R}}\right)
$$
to transform to a frame that is co-moving with the shock.


The veclocities are altered in the co-moving frame, but the other quantities remain unchanged,
$$
\hat{u}_*=u_*-S_3, \hat{u}_{\mathrm{R}}=u_{\mathrm{R}}-S_3.
$$
Since the wave is stationery in the co-moving frame, the conditions read,
$$
\begin{aligned}
\rho_* \hat{u}_* & =\rho_{\mathrm{R}} \hat{u}_{\mathrm{R}} \\
\rho_* \hat{u}_*^2+p_* & =\rho_{\mathrm{R}} \hat{u}_{\mathrm{R}}^2+p_{\mathrm{R}} \\
\hat{u}_*\left(\hat{E}_*+p_*\right) & =\hat{u}_{\mathrm{R}}\left(\hat{E}_{\mathrm{R}}+p_R\right)
\end{aligned}
$$

These quantities can be used to calculate the Mach numbers,
$$
M_R = u_R/a_R,\quad M_S=S_3/a_R
$$
where $M_R$ is the Mach number of the flow ahead of the shock in the original frame, and $M_S$ is the shock Mach number.
The resulting fluid speed is
$$
S_3=u_{\mathrm{R}}+a_R \sqrt{\left(\frac{\gamma+1}{2 \gamma}\right)\left(\frac{p_z}{p_{\mathrm{R}}}\right)+\left(\frac{\gamma-1}{2 \gamma}\right)}.
$$

The particle velocity behind the shockwave is,
$$
u_*=\left(1-\rho_{\mathrm{R}} / \rho_*\right) S_3+u_{\mathrm{R}} \rho_{\mathrm{R}} / \rho_*
$$

## The Riemann Problem

The Riemann problem has initial conditions,

$$
\mathbf{U}(x, 0)=\mathbf{U}^{(0)}(x)= \begin{cases}\mathbf{U}_{\mathrm{L}} & \text { if } x<0 \\ \mathbf{U}_{\mathrm{R}} & \text { if } x>0\end{cases}.
$$

The domain is taken to be the $\mathbb R$ and $t>0$.

The solution to the Riemann problem has 3 waves, one for each eigen-value.


The 3 waves bound 4 constant states labeled $W_L$, $W_{*L}$,$W_{*R}*$, $W_R$, where *Star Left* ($W_{*L}$) is the left side of the star region and *Star Right* ($W_{*L}$) is the right side.
The linear middle wave is a contact discontintuity, while the left and right (non-linear) waves are shock (discontinuous) or rarefaction (smooth) waves.

Since there is a contact discontintuity in the Star Region, we know that $p_{*L}=p_{*R}$ and $u_{*L}=u_{*R}$, we can therefore use this knowledge with the Rankine-Hugonoit conditions to join the left and right region.

Doing this requires solving,

$$
f\left(p, \mathbf{W}_{\mathrm{L}}, \mathbf{W}_{\mathrm{R}}\right) \equiv f_{\mathrm{L}}\left(p, \mathbf{W}_{\mathrm{L}}\right)+f_{\mathrm{R}}\left(p, \mathbf{W}_{\mathrm{R}}\right)+\Delta u=0, \Delta u \equiv u_{\mathrm{R}}-u_{\mathrm{L}}
$$

where the function $f_{\mathrm{L}}$ is given by

$$
f_{\mathrm{L}}\left(p, \mathbf{W}_{\mathrm{L}}\right)= \begin{cases}\left(p-p_{\mathrm{L}}\right)\left[\frac{A_{\mathrm{L}}}{p+B_{\mathrm{L}}}\right]^{\frac{1}{2}} & \text { if } p>p_{\mathrm{L}}(\text { shock }) \\ \frac{2 a_{\mathrm{L}}}{(\gamma-1)}\left[\left(\frac{p}{p_{\mathrm{L}}}\right)^{\frac{\gamma-1}{2 \gamma}}-1\right] & \text { if } p \leq p_{\mathrm{L}} \text { (rarefaction) }\end{cases}
$$

the function $f_{\mathrm{R}}$ is given by

$$
f_{\mathrm{R}}\left(p, \mathbf{W}_{\mathrm{R}}\right)= \begin{cases}\left(p-p_{\mathrm{R}}\right)\left[\frac{A_{\mathrm{R}}}{p+B_{\mathrm{R}}}\right]^{\frac{1}{2}} & \text { if } p>p_{\mathrm{R}}(\text { shock ), } \\ \frac{2 a_{\mathrm{R}}}{(\gamma-1)}\left[\left(\frac{p}{p_{\mathrm{R}}}\right)^{\frac{\gamma-1}{2 \gamma}}-1\right] & \text { if } p \leq p_{\mathrm{R}} \text { (rarefaction), }\end{cases}
$$

and the data-dependent constants $A_{\mathrm{L}}, B_{\mathrm{L}}, A_{\mathrm{R}}, B_{\mathrm{R}}$ are given by

$$
\left.\begin{array}{l}
A_{\mathrm{L}}=\frac{2}{(\gamma+1) \rho_{\mathrm{L}}}, B_{\mathrm{L}}=\frac{(\gamma-1)}{(\gamma+1)} p_{\mathrm{L}} \\
A_{\mathrm{R}}=\frac{2}{(\gamma+1) \rho_{\mathrm{R}}}, B_{\mathrm{R}}=\frac{(\gamma-1)}{(\gamma+1)} p_{\mathrm{R}}
\end{array}\right\}
$$

where the solution for the particle velocity in the Star Region is,
$$
u_*=\frac{1}{2}\left(u_{\mathrm{L}}+u_{\mathrm{R}}\right)+\frac{1}{2}\left[f_{\mathrm{R}}\left(p_*\right)-f_{\mathrm{L}}\left(p_*\right)\right]
$$

The approach is normally to solve the pressure $p_*$ numerically, with $f(p)=0$ and then use that result to solve for $u_*$.

```julia
# Taken from https://github.com/ryarazi/ExactRiemannProblemSolver
using CairoMakie
using IJulia

include(joinpath(@__DIR__, "ExactEulerRiemannProblemSolver", "RiemannSolver.jl"))

N = 1000
xmin = -0.5
xmax = 0.5
x = collect(LinRange(xmin, xmax, N))
save_figs = true #if this is false the figures will use show(), otherwise they will be saved locally in 'figs' dir

#create directory for figures
if save_figs && !isdir("figs")
	mkdir("figs")
end

struct TestCase
	name::String
	t::Float64
	left::HydroStatus
	right::HydroStatus
end

function calculate_and_plot(test::TestCase)
	name = test.name
	
	println("Time for the test $name:")
	
	#dummy run to compile all code
	sample_riemann(x, test.t, test.left, test.right)
	
	@time profiles = sample_riemann(x, test.t, test.left, test.right)
	density = [status.rho for status in profiles]
	velocity = [status.u for status in profiles]
	pressure = [status.p for status in profiles]
	energy = pressure ./ density ./ (test.left.gamma - 1.)
	
    fig = Figure()
    ax = Axis(fig[1, 1], ylabel= "density", xlabel="x")
	lines!(ax, x, density)

    ax = Axis(fig[1, 2], ylabel= "velocity", xlabel="x")
	lines!(ax, x, velocity)

    ax = Axis(fig[2, 1], ylabel= "pressure", xlabel="x")
	lines!(ax, x, pressure)

    ax = Axis(fig[2, 2], ylabel= "energy", xlabel="x")
	lines!(ax, x, energy)

	IJulia.clear_output(true)
	
	display(fig)	
end

gamma = 7. / 5.

#The first 5 case are taken from Toro's book: (Table 4.1)
#Toro, E.F., Riemann Solvers and Numerical Methods for Fluid Dynamics, A Practical Introduction, 3nd ed., Springer, Berlin, 2009.
#cases 6-13 are takeb from:
#Kamm et. al. 'Enhanced Verification Test Suite for Physics Simulation Codes' SAND2008-7813 2009
```

```julia

#case 1: 
for t in range(0, 0.25, length=100)
	sod = TestCase("SodShockTube",  t, HydroStatus(1., 0., 1., gamma),
									  HydroStatus(0.125, 0., 0.1, gamma))
	calculate_and_plot(sod)
end
```

```julia
#case 2: 
for t in range(0, 0.15, length=100)
	einfeldt = TestCase("Einfeldt123",  t, HydroStatus(1., -2., 0.4, gamma),
									      HydroStatus(1., 2., 0.4, gamma))
	calculate_and_plot(einfeldt)
end
```

```julia
#case 3:
for t in range(0, 0.012, length=100)
    toro3 =  TestCase("ToroTest3",  t, HydroStatus(1., 0., 1000., gamma),
									   HydroStatus(1., 0., 0.1, gamma))
    calculate_and_plot(toro3)
end
```

$$
\partial_t u(x) \rightarrow\frac{u^{n+1}_i-\frac{1}{2}(u^{n}_{i-1}+u^{n}_{i-1})}{\Delta t} 
$$

$$
\partial_t u(x)\frac{1}{2} \rightarrow\frac{u^{n+1}_{i+1}-u^{n}_{i-1}}{\Delta x}
